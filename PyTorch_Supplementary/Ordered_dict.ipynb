### 기존 dict의 문제점 : for loop으로 dict돌릴때 unordered 되게 나왔었다!!


하지만, 지금은 dict해도 여전히 순서가 유지가 되도록 나와서 문제가 없어졌다... 하지만, 만약 과거 package들과 같이 쓴다면 호환성을 위해 `OrderedDict`를 쓰는 것이 좋다


### `OrderedDict`사용 예시

```python
from collections import OrderedDict
new_dict = OrderedDict()
new_dict['A'] = 1
new_dict['B'] = 2
new_dict['C'] = 3
print(new_dict)
for key, val in new_dict.items():
    print(key ,val)
```

```python
##ordered dict자체에서 정의하기 #dictionary를 wrapping해야함!
new_dict_2 = OrderedDict({'A':1, 'B':2, 'C':3})
```

```python
#참고로 dictinoary의 items를 뽑으면 loop돌릴 수 있다
new_dict_2.items() #이것을 loop돌리기 가능
```

```python
for i in new_dict:
    print(i)
for i in new_dict.items():
    print(i)
```

```python
dictionary = {'A':1, 'B':2,'C':3}
dictionary.items()
```

```python
for key,val in dictionary.items():
    print(key, val)
```
