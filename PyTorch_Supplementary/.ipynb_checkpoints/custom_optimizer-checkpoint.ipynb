## Custom optimizer
(참조 : http://mcneela.github.io/machine_learning/2019/09/03/Writing-Your-Own-Optimizers-In-Pytorch.html)
크게 뭐 하거나 그러지는 않을 것이다 하지만 이것들만 알고 있자.

1. **모든** optimizer은 (`optim.SGD` 같은 것들까지 다 포함해서)
    * must inherit from `optim.Optimizer` (have their methods and stuff)
        * `optim.Optimizer` requires two things
            * `params = model.parameters()` (i.e. what to optimize)
            * `defaults = {'lr':3, 'momenutm': 0.9', ...}` (i.e. dict형태로, 기본적인 값들
    * must have the `step()` method (실제로 step을 할때 써야하니)
이것들만 있으면 정의해서 할 수있다. 


2. 어쨋튼, custom optimizer은 `optim.Optimizer` 을 inherit하기때문에, **모든 optimizer은 비슷하게 사용가능** (save, load도 그래서 가능한 것으로 보임)

실제로, prebuilt optimizers도 저 custom optimizer과 같다.

밑의 SGD optimizer 의 source code 를 보면, `__init__`에서 `super` 을 써서 `Optimizer`을 inherit하면서 `step`함수를 정의해 놓았다는 것을 알 수 있다
