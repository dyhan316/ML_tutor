#### AUC ROC example

(https://scikit-plot.readthedocs.io/en/stable/metrics.html#scikitplot.metrics.plot_roc)

```python
# The usual train-test split mumbo-jumbo
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
nb = GaussianNB()
nb.fit(X_train, y_train)
predicted_probas = nb.predict_proba(X_test)

# The magic happens here
import matplotlib.pyplot as plt
import scikitplot as skplt
skplt.metrics.plot_roc(y_test, predicted_probas)
plt.show()
```

#### Trying to implement it into a bianry classification thing (not one-hot encoded)

==================

`plot_sth_sth( target, prediction)` 

where, 
* `target` : N array, (not one-hot encoded)
* `prediction` : NxC array (i.e. probability array of being in each class


==================

, , (where now)

* `target` : N array (0 or 1) => YES should work 
* `prediction` : N array (% of being 1)

Therefore, to be able to use Binary Classification with this module, we need to change
prediction to array of type `Nx2`, with each element of the array being `(1-pred , pred)`



```python
#importing sample binary classification data
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
import numpy as np 
X, y = load_breast_cancer(return_X_y=True)

clf = LogisticRegression(solver="liblinear", random_state=0).fit(X, y)

pred = clf.predict_proba(X)[:,1] #only take the second ones (i.e. chance of being 1)
```

```python
##looking at tsom estuff 
print(X.shape, y.shape, pred.shape)
X[17:20], y[17:20], pred[17:20] #can see that y and pred are similar
```

```python
##trying to implement binary classification AUCROC plot 

import scikitplot as skplt
import matplotlib.pyplot as plt 
import sklearn

##trying with sth new
y = np.array([1,0,0,1])
pred = np.array([0.99, 0.9,0.3 ,0.9]) #pred = np.array([0.99, 0.01,0.3 ,0.7])
#print(pred.shape)


##calculating the thing 
print(sklearn.metrics.roc_auc_score(y, pred))
print(sklearn.metrics.roc_auc_score(y, pred, average='micro'))
print(sklearn.metrics.roc_auc_score(y, pred, max_fpr=0.5)) # don't use this (yet)

#micro ROC      =>    
#     multiclass   sklearn binary class    
#   , sklearn   ? 
#also partial AUC thing?


print("prediction values   print npy ! (or dataframe, csv), also, implement partial AUC")



# ! (from N arr prob => Nx2 prob array)
pred_arr = np.array([[1-pred_i, pred_i] for pred_i in pred])  #pred array



##plotting the thing
skplt.metrics.plot_roc(y, pred_arr, 
                       title = "AD CN ROC Curves", #put in task names here
                      figsize = (6,6), title_fontsize="large",
                       plot_micro = False, plot_macro = False, 
                      classes_to_plot=[1])  #done becasue we only need to plot one of the two AUC (they're the same)
                    # class   y  0,1   
                    #y  'yes' 'no'  [0,'yes'] ? 
                    #classes_to_plot=["AD", "CN"])

print("prediction values   print npy ! (or dataframe, csv)")
#also, accuracy, recall, bias   print? 
#or, pattial AUC  ? 


plt.legend(['class name'])#,'random'])

# !!! 

#skplt.metrics.plot_precision_recall_curve(y, pred)
```
